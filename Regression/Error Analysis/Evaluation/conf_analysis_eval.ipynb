```python
# File to evaluate the confidence analysis 

import pandas as pd
import numpy as np

FOLDER_PATH = '../../../Datasets/Evaluations/Regression/Error_Analysis/confidence_analysis/'
```

```python
#Import all CSV Files from a specific folder

import os
import csv

def import_csv_files(folder_path):
    files = os.listdir(folder_path)
    
    csv_files = [file for file in files if file.endswith('.csv')]
    file_paths = []
    for csv_file in csv_files:
        file_paths.append(os.path.join(folder_path, csv_file))

    csv_files_stripped = [file[:-4] for file in csv_files]
    return file_paths, csv_files_stripped

file_paths, csv_files = import_csv_files(FOLDER_PATH)

```

```python
def extract_confidence(value):
    return float(value.split(':')[1].strip().replace('%', ''))

def categorize_deviation(deviation):
    for i, threshold in enumerate(thresholds, start=1):
        if deviation <= threshold:
            return f'Up to {i}x STD'
    return f'Greater than {len(thresholds)}x STD'


for i in range(len(file_paths)):
    df_data = pd.read_csv(file_paths[i])
    print(csv_files[i], ":")
    #print(df_data.head())
    df_data['analysis_value'] = df_data['analysis'].apply(extract_confidence)
    result = df_data.groupby('error')['analysis_value'].agg(['mean', 'std', 'var'])
    result2 = df_data['analysis_value'].agg(['mean', 'std', 'var'])

    print(result)
    print("overall:")
    print(result2)
    print()
    df_data['deviation'] = (df_data['selling_price'] - df_data['prediction']).abs()
    std_dev = df_data['deviation'].std()
    thresholds = [0.5 * std_dev, 1 * std_dev, 1.5 * std_dev, 2 * std_dev]
    df_data['deviation_category'] = df_data['deviation'].apply(categorize_deviation)
    average_confidences = df_data.groupby('deviation_category')['analysis_value'].mean()
    print(average_confidences)

    print("\n")
```
