```python
#This file is used to analyze the datasets generated from the structured explanations

import pandas as pd
import numpy as np

FOLDER_PATH = '../../../Datasets/Evaluations/Regression/Error_Analysis/structured_unstructured/'
```

```python
import os
import csv

#Import the csv files from a given Folder
def import_csv_files(folder_path):
    files = os.listdir(folder_path)
    
    csv_files = [file for file in files if file.endswith('.csv')]
    file_paths = []
    for csv_file in csv_files:
        file_paths.append(os.path.join(folder_path, csv_file))

    csv_files_stripped = [file[:-4] for file in csv_files]
    return file_paths, csv_files_stripped

file_paths, csv_files = import_csv_files(FOLDER_PATH)

```

```python
import json
def eval(df):
    df_columns = pd.DataFrame()
    df_columns["attribute"] = None
    df_columns["importance"] = None
    index = 0

    for i in range(len(df)):
        js = df.loc[i, "structured_analysis"]
        js_json = json.loads(js)
        for k in js_json:
            #print(index)
            df_columns.loc[index, ['attribute']] = k['attribute']
            df_columns.loc[index, ['importance']] = k['importance']
            index += 1
    return df_columns

def eval_dfcolumns(df_columns):
    #print('Count:')
    #print(df_columns['attribute'].value_counts())
    df_columns['importance'] = df_columns['importance'].astype(float)
    #print('Mean:')
    #print(df_columns.groupby('attribute')['importance'].mean())
    #print('Variance:')
    #print(df_columns.groupby('attribute')['importance'].var())


    df_return = pd.DataFrame(columns=['attribute', 'count', 'mean', 'variance'])
    df_return['attribute'] = df_columns['attribute'].value_counts().index

    for i in range(len(df_return)):
        df_values = df_columns[df_columns['attribute'] == df_return.loc[i, 'attribute']]
        df_return.loc[i, 'count'] = len(df_values)
        df_return.loc[i, 'mean'] = df_values['importance'].mean()
        df_return.loc[i, 'variance'] = df_values['importance'].var()
        df_return.loc[i, 'std'] = df_values['importance'].std()

        

    df_return.round(2)
    print(df_return)
    return df_return
    
        
for i in range(len(file_paths)):
    df_data = pd.read_csv(file_paths[i])
    print(csv_files[i], ":")

    df_columns = eval(df_data)
    df_return = eval_dfcolumns(df_columns)

    print("\n")



```
