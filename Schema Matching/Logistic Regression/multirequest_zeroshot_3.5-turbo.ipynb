```python

import pandas as pd

OUTPUT_PATH = f"../../Datasets/Evaluations/Schema Matching/logistic_regression.csv"
TEST_CORRESPONDENCES_PATH = "../../Datasets/t2d_sm_nh/test/test_correspondences_sample_logistic.csv"
TRAIN_CORRESPONDENCES_PATH = "../../Datasets/t2d_sm_nh/train/train_correspondences_sample.csv"

```

```python
# Read the CSV file into a DataFrame
df_test = pd.read_csv(TEST_CORRESPONDENCES_PATH)
df_train = pd.read_csv(TRAIN_CORRESPONDENCES_PATH)

df_test = df_test.drop("prompt", axis=1)
df_test = df_test.drop("prompt_name", axis=1)
df_train = df_train.drop("prompt", axis=1)
df_train = df_train.drop("prompt_name", axis=1)

#print(df_test.head())
print(df_train.head())
```

```python
last = ""
table_A = ""
table_B = ""

for i in range(0, len(df_test)):
    if last != df_test["table_name"].iloc[i]:
        last = df_test["table_name"].iloc[i]
        table_A = df_test["table_A"].iloc[i]
        table_B = df_test["table_B"].iloc[i]
        print(i)
    else:
        df_test.loc[i, 'table_A'] = table_A
        df_test.loc[i, 'table_B'] = table_B

last = ""
table_A = ""
table_B = ""

for i in range(0, len(df_test)):
    if last != df_train["table_name"].iloc[i]:
        last = df_train["table_name"].iloc[i]
        table_A = df_train["table_A"].iloc[i]
        table_B = df_train["table_B"].iloc[i]
        print(i)
    else:
        df_train.loc[i, 'table_A'] = table_A
        df_train.loc[i, 'table_B'] = table_B
```

```python
df_train['y_true'] = df_train['y_true'].replace(True, 1).replace(False, 0)
df_test['y_true'] = df_test['y_true'].replace(True, 1).replace(False, 0)
print(df_train['table_B'][1])


```

```python
df_train['tables_combined'] = df_train['table_A'] + " " + df_train['table_B']
df_test['tables_combined'] = df_test['table_A'] + " " + df_test['table_B']
df_train['tables_combined'] = df_train['tables_combined'].astype(str)
df_test['tables_combined'] = df_test['tables_combined'].astype(str)

df_train['table_A'] = df_train['table_A'].astype(str)
df_train['table_B'] = df_train['table_B'].astype(str)
df_test['table_A'] = df_test['table_A'].astype(str)
df_test['table_B'] = df_test['table_B'].astype(str)

```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
from sklearn.preprocessing import MaxAbsScaler
from sklearn.decomposition import TruncatedSVD

# Initialize the TF-IDF Vectorizer for each independent column
tfidf_vectorizer_A = TfidfVectorizer()
tfidf_vectorizer_B = TfidfVectorizer()
tfidf_vectorizer_table = TfidfVectorizer()

# Fit on the training data and transform both training and testing data for each column
tfidf_A_train = tfidf_vectorizer_A.fit_transform(df_train['column_table_A'])
tfidf_B_train = tfidf_vectorizer_B.fit_transform(df_train['column_table_B'])
tfidf_table_train = tfidf_vectorizer_table.fit_transform(df_train['tables_combined'])

tfidf_A_test = tfidf_vectorizer_A.transform(df_test['column_table_A'])
tfidf_B_test = tfidf_vectorizer_B.transform(df_test['column_table_B'])
tfidf_table_test = tfidf_vectorizer_table.transform(df_test['tables_combined'])

X_train = hstack([tfidf_A_train, tfidf_B_train, tfidf_table_train])
X_test = hstack([tfidf_A_test, tfidf_B_test, tfidf_table_test])


y_train = df_train['y_true']
y_test = df_test['y_true']

#took paramter values that made sence / roughly fitted on training data until it stopped only responding with false values
model = LogisticRegression(C=1, max_iter=1000, class_weight={0: 1, 1: 1.5})

model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)

y_pred_train = model.predict(X_train)

accuracy_test = accuracy_score(y_test, y_pred_test)
conf_matrix_test = confusion_matrix(y_test, y_pred_test)
precision_test = precision_score(y_test, y_pred_test)
recall_test = recall_score(y_test, y_pred_test)
f1_test = f1_score(y_test, y_pred_test)


accuracy_train = accuracy_score(y_train, y_pred_train)
conf_matrix_train = confusion_matrix(y_train, y_pred_train)
precision_train = precision_score(y_train, y_pred_train)
recall_train = recall_score(y_train, y_pred_train)
f1_train = f1_score(y_train, y_pred_train)

print("Testing Set Results:")
print(f"Accuracy: {accuracy_test}")
print(f"Precision: {precision_test}")
print(f"Recall: {recall_test}")
print(f"F1 Score: {f1_test}")
print(f"Confusion Matrix:\n{conf_matrix_test}")

print("\nTraining Set Results:")
print(f"Accuracy: {accuracy_train}")
print(f"Precision: {precision_train}")
print(f"Recall: {recall_train}")
print(f"F1 Score: {f1_train}")
print(f"Confusion Matrix:\n{conf_matrix_train}")

df_test['y_pred'] = y_pred_test
```

```python
import numpy as np
# Combine feature names
feature_names = np.concatenate([
    tfidf_vectorizer_A.get_feature_names_out(),
    tfidf_vectorizer_B.get_feature_names_out()
])

# Extract coefficients
coefficients = model.coef_[0]

# Create a mapping from feature names to coefficients
feature_weights = dict(zip(feature_names, coefficients))

import re

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove non-alphanumeric characters
    text = re.sub(r'\W+', ' ', text)
    # Return the preprocessed text
    return text

def get_token_weights(text, vectorizer, feature_weights):
    # Apply preprocessing to match the vectorizer's preprocessing
    preprocessed_text = preprocess_text(text)
    # Split the preprocessed text into tokens
    tokens = preprocessed_text.split()
    # Retrieve and format the weights for each token found in feature_weights
    weights_str = ", ".join([f"{token}: {feature_weights.get(token, 0):.4f}" for token in tokens if token in feature_weights])
    return weights_str

df_test['token_weights'] = df_test.apply(lambda row: get_token_weights(row['column_table_A'], tfidf_vectorizer_A, feature_weights) + ", " +
                                                      get_token_weights(row['column_table_B'], tfidf_vectorizer_B, feature_weights), axis=1)
```
