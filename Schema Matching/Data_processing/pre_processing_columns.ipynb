```python
# File to pre-process the columns of the Testset for the Zero-Shot Baseline Analysis

import os
import pandas as pd

# Load the .env file

DATA_PATH = "../../Datasets/t2d_sm_nh/test/test_correspondences.csv"
OUTPUT_PATH = "../../Datasets/t2d_sm_nh/test/test_correspondences_sample.csv"
DB_PEDIA_PATH = "../../Datasets/t2d_sm_nh/test/dbpedia_tables/{name}.csv"
WEBTABLES_PATH = "../../Datasets/t2d_sm_nh/test/webtables/{name}.csv"
df = pd.read_csv(DATA_PATH)
print(df['label'].value_counts())
#sample_df = df.sample(n=450, random_state=41)
sample_df = df.head(450)
sample_df.reset_index(drop=True, inplace=True)
df = sample_df

# Read the CSV file into a DataFrame
df.rename(columns={'label': 'y_true'}, inplace=True)
df.rename(columns={'Unnamed: 0': 'table_index'}, inplace=True)

df['y_pred'] = None
df['prompt_name'] = None
df['prompt'] = None
df['column_table_A'] = None
df['column_table_B'] = None

df['y_true'] = df['y_true'].replace({0: "False", 1: "True"})

import random
for i in range (len(df)):
    row = df.iloc[i]
    table_name = row['table_name']
    table_path = WEBTABLES_PATH.format(name=table_name)
    pedia_path = DB_PEDIA_PATH.format(name=table_name)
    table = pd.read_csv(table_path)
    table_index = row['column_index_left']    
    pedia = pd.read_csv(pedia_path)
    pedia_index = row['column_index_right']
    
    column_table_A = table[str(table_index)]
    column_table_A_sample = column_table_A.sample(n=5, random_state=41)
    column_table_A_string = ' | '.join(column_table_A_sample.astype(str))
    column_table_B = pedia[str(pedia_index)]
    column_table_B_sample = column_table_B.sample(n=5, random_state=41)
    column_table_B_string = ' | '.join(column_table_B_sample.astype(str))
    df.loc[i, 'column_table_A'] = column_table_A_string
    df.loc[i, 'column_table_B'] = column_table_B_string
    print(i)

print(df.head())



```
