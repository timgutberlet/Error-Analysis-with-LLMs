```python

import pandas as pd

OUTPUT_PATH = f"../../Datasets/Evaluations/Schema Matching/logistic_regression.csv"
TEST_CORRESPONDENCES_PATH = "../../Datasets/t2d_sm_nh/test/test_correspondences_sample.csv"
TRAIN_CORRESPONDENCES_PATH = "../../Datasets/t2d_sm_nh/train/train_correspondences_sample.csv"

```

```python
# Read the CSV file into a DataFrame
df_test = pd.read_csv(TEST_CORRESPONDENCES_PATH)
df_train = pd.read_csv(TRAIN_CORRESPONDENCES_PATH)

df_test = df_test.drop("prompt", axis=1)
df_test = df_test.drop("prompt_name", axis=1)
df_train = df_train.drop("prompt", axis=1)
df_train = df_train.drop("prompt_name", axis=1)

print(df_test.head())
print(df_train.head())
```

```python
last = ""
table_A = ""
table_B = ""

for i in range(0, len(df_test)):
    if last != df_test["table_name"].iloc[i]:
        last = df_test["table_name"].iloc[i]
        table_A = df_test["table_A"].iloc[i]
        table_B = df_test["table_B"].iloc[i]
        print(i)
    else:
        df_test.loc[i, 'table_A'] = table_A
        df_test.loc[i, 'table_B'] = table_B

last = ""
table_A = ""
table_B = ""

for i in range(0, len(df_test)):
    if last != df_train["table_name"].iloc[i]:
        last = df_train["table_name"].iloc[i]
        table_A = df_train["table_A"].iloc[i]
        table_B = df_train["table_B"].iloc[i]
        print(i)
    else:
        df_train.loc[i, 'table_A'] = table_A
        df_train.loc[i, 'table_B'] = table_B
```

```python
df_train['y_true'] = df_train['y_true'].replace(True, 1).replace(False, 0)
df_test['y_true'] = df_test['y_true'].replace(True, 1).replace(False, 0)
print(df_train.head())
print(df_test.head())
```

```python
df = df_train
df['tables_combined'] = df['table_A'] + "_" + df['table_B']


new_df = df[['y_true', 'column_table_A', 'column_table_B']]
df = new_df

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# One-hot encode the string features
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df.drop('y_true', axis=1))

#Class weight Parameter - String balanced mitgeben


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(encoded_features, df['y_true'], test_size=0.2, random_state=42)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)


print("Trainset:")
# Predict on the testing set
y_pred = model.predict(X_train)

# Evaluate the model
accuracy = accuracy_score(y_train, y_pred)
conf_matrix = confusion_matrix(y_train, y_pred)

print("Accuracy of the model:", accuracy)
print("Confusion Matrix:\n", conf_matrix)


print("Testset:")
# Predict on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy of the model:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```
